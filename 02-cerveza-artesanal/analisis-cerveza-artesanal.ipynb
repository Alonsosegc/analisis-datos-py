## Ejercicio 2: Fabricación de cerveza artesanal

```python
import numpy as np #Importa la librería NumPy.
```

```python
array = np.loadtxt("alcohol_cerveza.csv", delimiter = ",", dtype= float, encoding= "utf-8", skiprows=1) 
#Carga los datos del archivo csv en una variable llamada array
#Se utiliza skiprows para saltarse la primera fila de los datos que tiene un valor de 0, que pudo haber sido usado como un dato para separar o dar inicio a los otros.
```

```python
array
```

```python
media = np.mean(array)   #Calcula la media de los datos
(f"La media del porcentaje de alcohol es de: {media:.3f}%")
```

```python
desviacion = np.std(array) #Calcula la desviación estándar de los datos
(f"La desviación estándar del porcentaje de alcohol es de: {desviacion:.3f}%")
```

```python
lic = media - (3* desviacion) #Calcula el límite inferior de control (LIC)
(f"El límite inferior de control del porcentaje de alcohol es de: {lic:.3f}%")
```

```python
lsc = media + (3*desviacion) #Calcula el límite superior de control (LSC)
(f"El límite superior de control del porcentaje de alcohol es de: {lsc:.3f}%")
```

```python
fuera_limites = array[(array <lic) | (array >lsc)] #obtener los valores fuera de los límites de control
```

```python
(f"Los datos que están fuera de los límites de control son: {fuera_limites}")
```
