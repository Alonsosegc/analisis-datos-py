```python
import pandas as pd #importamos pandas para poder utilizar sus funciones dentro de nuestro trabajo
```

```python
astro = pd.read_csv("resumen_resultados_astronautas.csv", sep= ",", encoding = "UTF-8") #le damos nombre a nuestro df y lo relacionamos con lo que está dentro del archivo abierto
```

```python
astro.head() #hacemos este paso para evidenciar que se cargo bien el documento y así también conocer sus columnas
```

```python
astro.dtypes #Ejecuto este comando para conocer en que datos está cada columna, así saber si necesito convertir alguna cuando quiera ejecutar alguna operatoria o así.
```

```python
astro["altura"]= astro["altura"]/100 #Convertimos el valor en metros dividiendo por 100.
astro.head()
```

```python
astro["IMC"] = (astro["peso"]/(astro["altura"]*astro["altura"])).round(1) #Utilizamos la fórmula IMC=m/h*h e incluimos la columna IMC
```

```python
astro.head()
```

```python
astro["promedio_evaluaciones"] = (astro["evaluacion_fisica"]+astro["evaluacion_psicologica"]+astro["evaluacion_adaptativa"]+astro["evaluacion_medica"])/4
#Creamos la columna de promedio evaluaciones
```

```python
astro2 = astro[(astro["IMC"]>=18.5) & (astro["IMC"]<= 24.9)] #Se filtra solo los astronautas que cumplen con la condición solicitada de los límites de IMC
astro2
```

```python
astro2= astro[(astro["IMC"]>=18.5) & (astro["IMC"]<= 24.9) & (astro["promedio_evaluaciones"]>= 87)]
#El promedio de las 4 evaluaciones debe ser mayor o igual a 87. Filtra el DataFrame resultante y agrega esta condición al filtro 
#aplicado en el paso anterior.

astro2
```

```python
astro2= astro[(astro["IMC"]>=18.5) & (astro["IMC"]<= 24.9) & (astro["promedio_evaluaciones"]>= 87)].reset_index()
#Reinicia el índice del DataFrame resultante.
astro2
```

```python
#Modifica el valor de la columna "califica", para que diga "Si" en todas las filas del DataFrame resultante.
astro2["califica"] = astro2["califica"].replace("No","Si") #
astro2
```

```python
#Exporta el DataFrame resultante a un archivo .csv separado por comas con el nombre "astronautas_calificados.csv".
astro2.to_csv("astronautas_calificados.csv")
```
